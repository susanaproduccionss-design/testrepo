```python
# TASK 3: Calculate the distances between a launch site to its proximities
def calculate_distance(lat1, lon1, lat2, lon2):
    # Radio de la Tierra en km
    R = 6373.0

    lat1, lon1 = np.radians(lat1), np.radians(lon1)
    lat2, lon2 = np.radians(lat2), np.radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    return R * c
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# Definir el sitio de lanzamiento central
launch_site_coords = [28.563197, -80.576820]

# Diccionario con los puntos de interés
proximities = {
    'Highway': [28.56335, -80.57085],
    'Railway': [28.57115, -80.58542],
    'City': [28.61200, -80.80788]
}

# Iterar sobre los puntos para calcular y dibujar
for name, coords in proximities.items():
    # 1. Calcular distancia
    distance = calculate_distance(launch_site_coords[0], launch_site_coords[1], coords[0], coords[1])
    
    # 2. Crear marcador con el texto de la distancia
    distance_marker = folium.Marker(
        coords,
        icon=DivIcon(
            icon_size=(150,30),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 10pt; color:#1565C0; width: 200px;"><b>{name}: {distance:.2f} KM</b></div>',
        )
    )
    site_map.add_child(distance_marker)
    
    # 3. Dibujar la línea conectora (PolyLine)
    line = folium.PolyLine(
        locations=[launch_site_coords, coords],
        weight=1.5,
        color='black' if name == 'City' else 'green', # Diferenciar la ciudad por color
        dash_array='5, 5' # Línea punteada para mejor visibilidad
    )
    site_map.add_child(line)

# Mostrar el mapa actualizado
site_map
```

```python
##1. ¿Están los sitios de lanzamiento cerca de los ferrocarriles?
Sí.

Observación: La distancia suele ser inferior a 1-2 km.

Explicación: Los cohetes Falcon 9 y sus componentes son masivos. El transporte por ferrocarril es la forma más eficiente y económica de trasladar las etapas del cohete desde la fábrica (en Hawthorne, CA o Texas) hasta el sitio de lanzamiento. Una vía férrea cercana minimiza el riesgo de daños durante el transporte terrestre final.

2. ¿Están los sitios de lanzamiento cerca de las autopistas?
Sí.

Observación: Se encuentran a una proximidad inmediata de carreteras principales o autopistas de acceso restringido.

Explicación: Se requiere un acceso rápido para el personal, los servicios de emergencia y el transporte de suministros diarios. Sin embargo, aunque están "cerca", suelen tener puntos de control de seguridad (checkpoints) a varios kilómetros de la plataforma para mantener un perímetro seguro.
```

```python
##3. ¿Están los sitios de lanzamiento cerca de la línea de costa?
Sí, de manera crítica.

Observación: Es la proximidad más estrecha (a menudo menos de 1 km).

Explicación: Esta es una decisión de seguridad fundamental. Al lanzar hacia el este (en Florida) o hacia el sur/oeste (en California), el cohete vuela inmediatamente sobre el océano. Si ocurre una anomalía o explosión en los primeros segundos del vuelo, los escombros caen al agua y no sobre zonas pobladas. Además, facilita la recuperación de los propulsores en las barcazas (droneships).

4. ¿Mantienen los sitios de lanzamiento cierta distancia de las ciudades?
Sí, mantienen una distancia considerable.

Observación: Las ciudades más cercanas (como Titusville o Lompoc) suelen estar a más de 15-25 km.

Explicación: Esto se debe a dos factores principales:

Riesgo de Explosión: En caso de un fallo catastrófico en la plataforma, la onda expansiva debe disiparse antes de llegar a zonas residenciales.

Contaminación Acústica: El ruido generado por los motores Merlin o Raptor durante el despegue alcanza niveles de decibelios que podrían dañar estructuras o la audición si las ciudades estuvieran más cerca.
```
